````python
# Extracted from the official web page
index = pd.MultiIndex.from_tuples([('bird', 'falcon'),
                                   ('bird', 'parrot'),
                                   ('mammal', 'lion'),
                                   ('mammal', 'monkey')],
                                  names=['class', 'name'])
columns = pd.MultiIndex.from_tuples([('speed', 'max'),
                                     ('species', 'type')])
df = pd.DataFrame([(389.0, 'fly'),
                   ( 24.0, 'fly'),
                   ( 80.5, 'run'),
                   (np.nan, 'jump')],
                  index=index,
                  columns=columns)
````

````python
df  # Exist  two levels class and name we want drop one level name
````

````python
df.reset_index(level='name')
````

````python
df.reset_index() # Remove all levels
````

````python
df['speed'] # Indexing 
````

````python
df.reset_index(level='class') # Note that a additional column (class) has been added to the datadrame
````

````python
df.reset_index(level='class', drop=True)  # Whith this not is included
# The level are the number of rows of the multiindex for instance look the table...
# Speed and Species are in the level(0) and max and type are in the level(1).
````

````python
df.reset_index(level='class', col_level=0) #Put class at level 0 by default is placed in the top Level
# Note that in the second level (1) of the class there is not text therefore we can put with fill_col='text'
````

````python
df.reset_index(level='class', col_level=0, col_fill='New_text')
````

# Stack and unstack
Methods to indexing dataframe (could be useful to multiindexing)

## Unstack is the inverse operation.


# Stack is "apilado" related with long

# Unstack is "no-apilado" related with wide

# Stack is associated with 


````python
import pandas as pd
import numpy as np
header = pd.MultiIndex.from_product([['Year 1994','Year 1995'],['Growth','Population']])
info=([[0.01,11,0.08,19],[0.013,12,0.022,13],[0.02,14,0.034,11],[0.03,15,0.056,17]])
df = pd.DataFrame(info,
                  index=['Colombia','Panamá','Chile','Argentina'],
                  columns=header)
#note Column header... have two levels years(level 0 ) and Growth and population(level 1 or -1(last))
df
````

````python
# We can stack this database...
# stack (apilar) the database is select one level(columns) and make panel with the level selected
df.stack(level=[0])   # This is very important note the following..
                    # Note the levels in columns names..
                    # remember that level beging in zero! therefore -1 is the most inner to data.
                    # change the level (1,-1,0) 
                    # It is important note that level could be a list....
                    # Note that the before database is index
````

````python
data.set_index(['Country','Year']).unstack()
````

````python
data.set_index(['Country','Year']).unstack().stack(level=1)
````

````python
data.set_index(['Country','Year']).unstack().stack(level=1).reset_index()
````
