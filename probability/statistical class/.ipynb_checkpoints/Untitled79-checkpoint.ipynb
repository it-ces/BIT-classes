# Variance properties
important the following properties, about variance:
* $var(x)>=0$
* $var(c)=0$
* $var(cx)=c^{2}var(x)$
* $var(x+c) = var(x)$
* $var(x) = E(x^{2}) - E^{2}(X)$
* $Var(x+y) \neq var(x)+ var(y)$
* the last properties is equal if the variables are independents.
proofs.


the last three properties are very useful in practice.

\begin{equation}
var(c) = \sum_{x} (x- \mu)^{2}f(x) \\
=(c-c)^² 1
\end{equation}
Note remember that $E(c) = c$

\begin{equation}
var(cx) = E[(cx-E(cx))^2] \\
 = E[(cx - cE(x))^2] \\
 =E[c^2(x-E(x))^2] \\
 = c^2E[(x-E(x))^2]
\end{equation}
thus is proved that $var(cx) = c^2var(x)$
this it is important:
\begin{equation}
var(x+c) = E[((x+c) - E(x+c))^2]\\
= E[(x+c - E(x) -c )^2 \\
= E[(x-E(x))^2] \\
= var(x) \\
\end{equation}

the following property it is important in the sense that allow us compute the variance more easily.

\begin{equation}
var(x)  = E[(x-E(x))^2] \\
= E[x^2 - 2xE(x) + E^2(x)] \\ 
= E(x^2) - 2E(x)E(x) + E^2(x) \\
= E(x^2) - 2E(x) + E^2(x) \\
= E(x^2) - E^2(x)
\end{equation}

the following property it is also very useful:
\begin{equation}
\mbox{if } y = x\\
var(x+y) = var(2x) = 4 var(x)\\
\mbox{ to the other case}\\
var(x) + var(y) = 2 var(x) \\
\mbox{in consequence} \\
var(x+y) \neq var(x)+var(y)
\end{equation}

Now we can see the case where $x \neq y$ 
\begin{equation}
var(x+y) = E(x+y)^2 - E²(x+y) \\
 = E(x^{2} + 2xy + y^2 ) - (E(x) + E(y))^2 \\
 = E(x^{2} + 2E(xy) + E(y^²) - E^2(y) \\
 = (E(x^2) - E^2(x)) + ( E(y^2) - E^2(y)) \\
 = Var(x) + var(y)
\end{equation}

## OLS
que have  a PMR, 
\begin{equation}
\mathbf{y} = \mathbf{\beta X + u}
\end{equation}

and we can get a Sampling approximation;
\begin{equation}
y_{i}  =  \hat{\beta_{0}} + \hat{\beta_{1}} x_{i}
\end{equation}

note that $u_{i} = y_{i} - \hat{y_{i}} $ and the OLS problem consist in 
$$ \min Z = \sum (u_{i})^{2}$$

the the problem could be stated as $Z(\beta_{0},\beta{1})$, for the first condition we can take that

\begin{equation}
\frac{\partial Z}{\partial \beta_{0} }= 0 \\
\frac{\partial Z}{\partial \beta_{0}} = \sum ( y_{i} - \beta_{0} - \beta{1}) = 0 \\
 = n\bar{y} - n\beta_{0} - n\beta_{1}\bar{x} = 0
\end{equation}
of this equation we can derive important features for OLS. For instance that:
\begin{equation}
\bar{y} = \hat{\beta_{0}} + \hat{\beta_{1}} \bar{x}
\end{equation}
this imply that the mean estimated is equal to the mean of data.

Note that $\hat{y_{i}} = \hat{\beta_{0}} + \hat{\beta_{1}}x_{1}$
and therefore 
\begin{equation}
\hat{y_i} - \bar{y} = \beta_{1}(x_{1} - \bar{x}) \\
\sum( \hat{y_i} - \bar{y})^{2}  = \beta_{1} \sum(x_{i} - \bar{x} ) \\
\end{equation}



# goodness of fit

The variance of data give us a measure of variability therefore we can tray explain the percentage of this data that we can predict with the model, this measure is called $r^{2}$
\begin{equation}
r^{2} =
\frac{\sum (\hat{y_{i}} - \bar{y} )^{2}}{\sum ( y_{i} - \bar{y})^{2}}
\end{equation}
however we can probe that $ 0 \leq r \leq 1$ it is easy if we take in mind that, 

\begin{equation}
\sum (y_{i} - \bar{y})^{2} = \\
\sum (y_{i} - \hat{y_i})^{2} + \sum (y_{i} - \bar{y})^{2}  = \\
SCT = SCE + SCR
\end{equation}
the we can see that the last equation that diving by $SCT$ that we claimed about $r^{2}$.
we also can view two important things, that $r^{2} = \frac{\hat{\beta_{1}}S_{x}}{S_{y}}$, where $S_{i}$ is the sample variance of the variable $i$.
\begin{equation}
SCR = SCT - SCE \\
SCR = SCT(1 - \frac{SCE}{SCT}) \\
SCR = SCT(1-r^{2}) \\
SCT = \sum(y_{i} - \bar{y}) (1-r^{2}) + r^{2}\sum({y_i} - \bar{y})^{2}
\end{equation}
