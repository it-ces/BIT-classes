# Variance properties
important the following properties, about variance:
* $var(x)>=0$
* $var(c)=0$
* $var(cx)=c^{2}var(x)$
* $var(x+c) = var(x)$
* $var(x) = E(x^{2}) - E^{2}(X)$
* $Var(x+y) \neq var(x)+ var(y)$
* the last properties is equal if the variables are independents.
proofs.


the last three properties are very useful in practice.

\begin{equation}
var(c) = \sum_{x} (x- \mu)^{2}f(x) \\
=(c-c)^² 1
\end{equation}
Note remember that $E(c) = c$

\begin{equation}
var(cx) = E[(cx-E(cx))^2] \\
 = E[(cx - cE(x))^2] \\
 =E[c^2(x-E(x))^2] \\
 = c^2E[(x-E(x))^2]
\end{equation}
thus is proved that $var(cx) = c^2var(x)$
this it is important:
\begin{equation}
var(x+c) = E[((x+c) - E(x+c))^2]\\
= E[(x+c - E(x) -c )^2 \\
= E[(x-E(x))^2] \\
= var(x) \\
\end{equation}

the following property it is important in the sense that allow us compute the variance more easily.

\begin{equation}
var(x)  = E[(x-E(x))^2] \\
= E[x^2 - 2xE(x) + E^2(x)] \\ 
= E(x^2) - 2E(x)E(x) + E^2(x) \\
= E(x^2) - 2E(x) + E^2(x) \\
= E(x^2) - E^2(x)
\end{equation}

the following property it is also very useful:
\begin{equation}
\mbox{if } y = x\\
var(x+y) = var(2x) = 4 var(x)\\
\mbox{ to the other case}\\
var(x) + var(y) = 2 var(x) \\
\mbox{in consequence} \\
var(x+y) \neq var(x)+var(y)
\end{equation}

Now we can see the case where $x \neq y$ 
\begin{equation}
var(x+y) = E(x+y)^2 - E²(x+y) \\
 = E(x^2 + 2xy + y^2 ) - (E(x) + E(y))^2 \\
 = E(x^² + 2E(xy) + E(y^²) - E^2(y) \\
 = (E(x^2) - E^2(x)) + ( E(y^2) - E^2(y)) \\
 = Var(x) + var(y)
\end{equation}
