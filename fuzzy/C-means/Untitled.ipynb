```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
import skfuzzy as fuzz
import array as arr
```

```python
x = np.arange(1,101)
data = np.reshape(x,[1,100])
conjuntos = 3
m = 2.0 # measure the abrupt change of a fuzzy set to another
error = 1e-9
max_iter = 20
centros, u, u0, d, jm, p, fpc = fuzz.cmeans(data, conjuntos, m, error, max_iter)
print("centroids")
print(centros)
plt.title("Ejemplo de c-means")
plt.xlabel("varpi")
plt.ylabel("membership")
# u is composed of three fuzzy sets with the degree of memembshersip of each of one its elements
# arange works as inrange excluding the last number
for i in np.arange(conjuntos):
    plt.plot(x, u[i])
```
